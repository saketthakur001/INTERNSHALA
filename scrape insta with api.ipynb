```python
import json
import os

def extract_comment_details(comments, media_info, json_dump_loc):
    """
    Extract key details from a list of comments, merge with media info, 
    and save them to a JSON file.

    Parameters:
    - comments: List of Comment objects.
    - media_info: Dictionary containing media post details.
    - json_dump_loc: Location of the JSON file to dump the data.

    Returns:
    - A list of dictionaries, each containing key details from a comment merged with media info.
    """
    details = []

    for comment in comments:
        detail = {
            'pk': comment.pk,
            'text': comment.text,
            'username': comment.user.username,
            'full_name': comment.user.full_name,
            'profile_pic_url': comment.user.profile_pic_url,
            'created_at_utc': comment.created_at_utc,
            'content_type': comment.content_type,
            'status': comment.status,
            'has_liked': comment.has_liked,
            'like_count': comment.like_count,
            'post_id': media_info['id']
        }

        # Merge comment details with media_info
        merged_data = {**detail, **media_info}
        details.append(merged_data)

    # Check if file exists, if so, load the existing data and append new data
    if os.path.exists(json_dump_loc):
        with open(json_dump_loc, 'r') as file:
            existing_data = json.load(file)
        existing_data.extend(details)
    else:
        existing_data = details

    # Write the combined data back to the file
    with open(json_dump_loc, 'w') as file:
        json.dump(existing_data, file, indent=4)

    return details
```
