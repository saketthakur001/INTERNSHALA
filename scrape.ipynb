```python
# Your existing imports
from selenium import webdriver
from selenium.webdriver.chrome.options import Options as chromeOptions
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import random
import os
import time
import re

# Additional imports
from selenium.webdriver.common.action_chains import ActionChains

# Your existing setup
options = webdriver.ChromeOptions()
options.add_argument("--user-data-dir=C:\\Temp\\ChromeProfile")
driver = webdriver.Chrome(options=options)

# Navigate to the Instagram page
driver.get('https://www.instagram.com/hdfcbank/')

# Wait for the posts to load
time.sleep(5)

# Find the first 6 posts
posts = driver.find_elements_by_class_name('v1Nh3')[:6]
posts = driver.find_elements(By.CLASS_NAME, 'v1Nh3')[:6]


# For each post
for i, post in enumerate(posts):
    print(f"Post {i+1}:")
    
    # Click on the post
    post.click()
    
    # Wait for the post to load
    time.sleep(5)
    
    # Find the 'Load more comments' button
    load_more_comments_button = driver.find_elements_by_xpath("//button[contains(text(), 'Load more comments')]")
    
    # While the 'Load more comments' button is visible
    while load_more_comments_button:
        # Click on the 'Load more comments' button
        load_more_comments_button[0].click()
        
        # Wait for more comments to load
        time.sleep(5)
        
        # Find the 'Load more comments' button again
        load_more_comments_button = driver.find_elements_by_xpath("//button[contains(text(), 'Load more comments')]")
    
    # Find all the comments
    comments = driver.find_elements_by_class_name('C4VMK')
    
    # For each comment
    for comment in comments:
        # Print the comment
        print(comment.text)
    
    # Close the post
    close_button = driver.find_element_by_xpath("//button[@aria-label='Close']")
    close_button.click()
    
    # Wait before moving on to the next post
    time.sleep(5)
```
